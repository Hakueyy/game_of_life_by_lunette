```python
1) Importer les librairies classiques : numpy, pandas, seaborn et matplolib
```

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
2) Lire le fichier Bank.csv 
```

```python
bank_doc = pd.read_csv("Bank.csv")
print(bank_doc)
```

```python
3) Créer trois listes(array) age,montant et lab (pour défaut)
```

```python
age = np.array(bank_doc['age du client'])
montant = np.array(bank_doc['montant du prêt'])
lab = np.array(bank_doc['défaut'])

print(age)
```

```python
4) Déterminer le nombre de non payés
```

```python
non_paye = sum(lab)
print(non_paye)
```

```python
5) Lister les montants non payés et calculer la somme non payés
```

```python
def indice_lab(l):
    ind = []
    for i in range(len(l)):
        if l[i]==1 : 
            ind.append(i)
    return ind

somme_non_payé = sum([montant[i] for i in indice_lab(lab)])
print(somme_non_payé)
print(len(age))
```

```python
6) Présenter graphiquement avec des petits circles(marker) de couleurs différentes (age, motant du prêt) selon la situation "défaut"
```

```python
plt.scatter(age[lab==0],montant[lab==0],color ="r",s=100)
plt.scatter(age[lab==1],montant[lab==1],color ="g",s=100)
plt.show()

# [lab==0] ne selectionne que les indices verifiant la condition lab==0 
```

```python
7) Présenter graphiquement avec les petits circles vertes (age,montant avec défaut=0) et dans le cas contraire rouge avec des labels
```

```python
plt.scatter(age[lab == 0], montant[lab == 0], color='g', label='Bon',s=100)
plt.scatter(age[lab == 1], montant[lab == 1], color='r', label='Pas Bon',s=100)
plt.legend() #affiche la légende
plt.grid() #affiche une grille
plt.xlabel('age') # nomme l'axe des abscisses 
plt.ylabel('montant') # nomme l'axe des ordonnées 
plt.show()
```

```python
8) Un client agé de 25an demande un prêt de 30000€. Peut on lui prêter cette somme? Vérification graphaique
```

```python
plt.scatter(age[lab == 0], montant[lab == 0], color='g', label='Bon',s=100)
plt.scatter(age[lab == 1], montant[lab == 1], color='r', label='Pas Bon',s=100)
plt.scatter(25,30000, color = 'b', label ='Inconnu', s=100)
plt.legend()
plt.grid()
plt.xlabel('age')
plt.ylabel('montant')
plt.show()

#le point inconnu est proche de points vert, on peut supposer qu'il est raisonnable de prêter de l'argent à cette personne
```

```python
9)Préparation des données. X= contient age et montant,  y= defaut
```

```python
X = np.zeros((len(age),2))
for i in range(len(age)):
    X[i,0]= age[i]
    X[i,1]= montant[i]

y = lab 

# fonction permettant de faire un tableau avec 2 colonnes [age,montant]



```

```python
10) Conception du modèl. Classification Knn
KNeighborsClassifier(n_neighbors=5,weights='uniform')
```

```python

model = KNeighborsClassifier(n_neighbors=5,weights='uniform')
model.fit(X,y)

# modèle qui prend une décision en fonction des voisins les plus proches. Le premiere argument définit le 
#nombre de voisins prit en compte. weight permet de faire varier le poids accordé à chaque voisins par exemple
#en fonction de la distance.
```

```python
11) Peut on prêter au client avec age=25an un prêt=30000€?   Vérification par modèle
```

```python
val =model.predict([[25,30000]])
print(val)

#val vaut 0 on peut donc prêter au client de 25 ans demandant 30000euros.
```

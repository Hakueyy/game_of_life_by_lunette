4) Créer trois listes(array) petal_length,petal_width et species

```python
petal_length = np.array(iris['petal_length'])
petal_width = np.array(iris['petal_width'])
species = np.array(iris['species'])

```

5) Tracer sur la même figure les nuages de point de "petal_width" en fonction de "petal_length". Les points seront de couleurs différentes selon leur catégorie respective dans "species" 

Les 2 couleurs suivantes sont à utiliser: vert pour setosa, rouge pour versicolor. Mettre une légande sur le  graphe 

```python
plt.scatter(petal_length[species==0],petal_width[species==0], color = 'g', label = 'setosa', s=100)
plt.scatter(petal_length[species==1],petal_width[species==1], color = 'r', label = 'versicolor', s=100)
plt.legend()
plt.grid()
plt.xlabel('petal_lenght')
plt.ylabel('petal_width')
plt.show()
```

6) Placer sur la figure précédente un point noir correspondant à un nouveau échantillon de fleur dont la pétale a les dimensions suivantes:
    largeur = 2,5 cm et longueur = 0,75 cm

```python
plt.scatter(petal_length[species==0],petal_width[species==0], color = 'g', label = 'setosa', s=100)
plt.scatter(petal_length[species==1],petal_width[species==1], color = 'r', label = 'versicolor', s=100)
plt.scatter(2.5,0.75, color = 'black', label = 'Inconnu', s = 100)
plt.legend()
plt.grid()
plt.xlabel('petal_lenght')
plt.ylabel('petal_width')
plt.show()
```

7) Séparer les données en X_train, X_test, Y_train, Y_test (80% pour entrainement)

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
X = np.array(list(zip(petal_length,petal_width)))
X_train,X_test,Y_train,Y_test = train_test_split(X,species, test_size=0.2)
```

8) Entrainement, test en utilisant l'algorithme SVM (SVC= pour classification,  SVR : pour regression)

```python
model = SVC( )
model.fit(X_train,Y_train)

new_y = model.predict(X_test)
print(new_y)
print(Y_test)
```

9) Vérifier le score "from sklearn.metrics import accuracy_score"

```python
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
```

```python

score = accuracy_score(Y_test,new_y)
print(score)
```

10) Tracer le plan (droite) de séparation entre les classes

```python




DecisionBoundaryDisplay.from_estimator(model,X, response_method='predict')
plt.scatter(petal_length[species==0],petal_width[species==0], color = 'g', label = 'setosa', s=100)
plt.scatter(petal_length[species==1],petal_width[species==1], color = 'r', label = 'versicolor', s=100)
plt.scatter(0.75,2.5, color = 'black', label = 'Inconnu', s = 100)
plt.legend()
plt.grid()
plt.xlabel('petal_lenght')
plt.ylabel('petal_width')
plt.show()

#DecisionBoundaryDisplay montre la frontière choisit par le modèle. 
```
